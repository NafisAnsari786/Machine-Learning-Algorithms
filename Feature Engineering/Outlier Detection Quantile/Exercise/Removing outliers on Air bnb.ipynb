**Removing Outliers from Air bnb new york data set**

```python
import pandas as pd
import numpy as np 
```

```python
df=pd.read_csv(r"F:\Machine Learning all Algorithms\Feature Engineering\Outlier Detection quantile\Exercise\AB_NYC_2019.csv")
df
```

```python
df.shape
```

```python
df.price.describe()
```

```python
min_thresold, max_thresold = df['price'].quantile([0.001, 0.999])
min_thresold, max_thresold
```

```python
df[df['price']<min_thresold].head()
```

```python
df[df['price']>max_thresold].head()
```

```python
df2=df[(df.price>min_thresold) & (df.price<max_thresold)]
df2
```

```python
df2.shape
```

```python
df2.sample(10)
```

```python
df2.describe()
```

```python
df2.to_csv("F:\Machine Learning all Algorithms\Feature Engineering\Outlier Detection quantile\Exercise\cleaned_air_bnb.csv")
```
