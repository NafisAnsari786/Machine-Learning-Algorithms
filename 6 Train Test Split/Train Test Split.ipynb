# <span style="color:Purple"> Training And Testing Available Data </span>
We have a dataset containing prices of used BMW cars. We are going to analyze this dataset and build a prediction function that can predict a price by taking mileage and age of the car as input. We will use sklearn train_test_split method to split training and testing dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"F:\Machine Learning all Algorithms\6 Train Test Split\carprices.csv")
df.head()
```

```python
plt.scatter(df['Mileage'], df['Sell Price($)'], color='red', marker='+')
```

```python
plt.scatter(df['Age(yrs)'], df['Sell Price($)'], color='blue', marker='*')
```

```python
X=df[['Mileage', 'Age(yrs)']]
y=df['Sell Price($)']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
len(X_train)        # based on the test size, 80% train data, 20% test data
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
model.score(X_test, y_test)
```
