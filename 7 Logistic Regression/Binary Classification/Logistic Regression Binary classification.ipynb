## Logistic Regression (Binary Classification)

## <span style="color:Purple"> Predicting if a person would buy life insurnace based on his age using logistic regression </span>

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"F:\Machine Learning all Algorithms\7 Logistic Regression\Binary Classification\insurance_data.csv")
df.head()
```

```python
plt.scatter(df['age'], df['bought_insurance'], marker='*', color='red')
plt.xlabel('Age')
plt.ylabel('Bought Insurance')
plt.show()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(df[['age']], df['bought_insurance'], test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
X_test
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
model.predict_proba(X_test) 

# Predicts the probability of data sample being in different classes, ( 0,1) 0 as- will not by insurance, 1 as- will by isnurance
# eg: Age 42, 18 % chance, they will not buy. 81 % chance, they will buy.
```
