# Gradient Descent

```python
import numpy as np
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0                         # M current and B current starts from 0 initially
    iterations = 1000
    n = len(x)                                  # n is the length of these data points
    learning_rate = 0.01

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr          # y=mx+b
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])        # we need to reduce the cost at each iteration
        md = -(2/n)*sum(x*(y-y_predicted))          # md is M derivative
        bd = -(2/n)*sum(y-y_predicted)              # bd is B derivative
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print("m {}, b{}, cost{}, iterations{}".format(m_curr, b_curr, cost, i))


x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])

gradient_descent(x,y)
```

## Plotting the Gradient Descent 

```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x,y):
    m_curr = b_curr = 0.0  # Initialize m and b as floats
    iterations = 10000
    learning_rate = 0.01
    n = float(len(x))  # Convert length of x to float for calculations
    
    # Plot the data points
    plt.scatter(x, y, color='red', marker='+', linewidth=5)
    
    for i in range(iterations):
        # Calculate predictions
        y_predicted = m_curr * x + b_curr  # y_predicted is an array, all numerical
        
        # Plot the line every 50 iterations
        if i % 50 == 0:
            plt.plot(x, y_predicted, color='green')
        
        # Calculate gradients
        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)
        
        # Update parameters
        m_curr -= learning_rate * md
        b_curr -= learning_rate * bd

    # Display the final plot
    plt.show()

# Define data points
x = np.array([1, 2, 3, 4, 5], dtype=float)  # Ensure x is float type
y = np.array([5, 7, 9, 11, 13], dtype=float)  # Ensure y is float type

# Call the function
gradient_descent(x, y)
```
